```python
import os
```

```python
import numpy as np

import pandas as pd
```

```python
PROJECT_DIR = r"C:\Users\Surface\Desktop\data-science-project\ML-Project\MLOps-Flight-Price-Prediction"
DATA_DIR = "notebook"
```

```python
def get_data(name):
	file_name = f"{name}.csv"
	file_path = os.path.join(PROJECT_DIR, DATA_DIR, file_name)
	return pd.read_csv(file_path)
```

```python
flights = get_data("flight_price")
flights
```

```python
flights.head()
```

```python
import pymongo
```

```python
# df should be converted into dict before we push it to mongodb

data = flights.to_dict(orient='records')
```

```python
DB_NAME = "Flight2"
COLLECTION_NAME = "flight2_data"
CONNECTION_URL = ""

# above, either remove your credentials or delete the mongoDB resource bofore pushing it to github.
```

```python
import pymongo
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]

rec = collection.insert_many(data)
```
